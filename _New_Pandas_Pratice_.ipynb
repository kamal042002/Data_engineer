# Extracts unique department names or for any columns

```python
def unique_data_from_table(Dataframe,column_name,output_file):
    unique=Dataframe[column_name].dropna().unique()
    pd.DataFrame(unique, columns=[f"Unique {column_name}"]).to_csv(output_file, index=False)
unique_data_from_table(Employees_data, "Department","Department_unique.csv")
```

# Write a Python script that:
# Reads a dataset containing Date and Revenue columns.
# Groups the data by month.
# Plots the monthly revenue trend using Seaborn line plot.¶


```python
orders=all_sheets_data["Orders"]
orders
```

```python
Revenue_data=orders[["Order Date","Quantity ordered new","Sales"]]
Revenue_data["Total_sales"]=Revenue_data["Quantity ordered new"]*Revenue_data["Sales"]
Revenue_data
```

```python
Revenue=Revenue_data[["Order Date","Total_sales"]]
Revenue
```

```python
monthly_sales=Revenue.groupby(Revenue["Order Date"].dt.strftime("%B"))["Total_sales"].sum().sort_values(ascending=True).reset_index()
monthly_sales
```

```python

sns.lineplot(data=monthly_sales,x="Order Date",y="Total_sales",markers="*")
sns.scatterplot(data=monthly_sales, x="Order Date", y="Total_sales", color="green", s=100)
```

# Product Category	wise sales

```python
product_Category=orders[["Product Category","Quantity ordered new","Sales","Customer Segment"]]
product_Category["Total_sales"]=product_Category["Quantity ordered new"]*product_Category["Sales"]
product_Category
```

```python
product_Category_sales=product_Category.groupby(["Product Category","Customer Segment"])["Total_sales"].sum().reset_index()
product_Category_sales
```

```python
ax=sns.barplot(data=product_Category_sales,x="Product Category",y="Total_sales",hue="Customer Segment")
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height):,}',            # add commas for readability
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=10)
```
